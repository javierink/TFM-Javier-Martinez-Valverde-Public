# Heatmap 

```python
sns.set()

data = df.pivot("user", "algorithm", "neg_mean_squared_error")
label_size = 14

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
hm = sns.heatmap(data, annot=True, linewidths=.5, ax=ax, cmap="Blues")
hm.set_xticklabels(hm.get_xmajorticklabels(), fontsize = label_size, rotation=45, horizontalalignment='right')
hm.set_yticklabels(hm.get_ymajorticklabels(), fontsize = label_size, rotation=0,horizontalalignment='right')
ax.set_xlabel('Modelo', fontsize=label_size , fontweight="bold")
ax.set_ylabel('Usuario', fontsize=label_size, fontweight="bold")

plt.savefig(plots_path + "heatmap_general.pdf")
```

# Lineplots of features importance

## Data preparation

```python
df = df[df.user == "General"].drop(["user", "neg_mean_squared_error"], axis=1).reset_index(drop=True)
```

```python
list_features = df.columns[1:]
list_algs = df.algorithm.unique().tolist()
```

```python
plt.rcParams.update({'font.family': 'monospace'})
sns.set(style="whitegrid",font_scale=3.4, rc={'figure.figsize':(20,10)})

plt.rcParams['axes.edgecolor'] = "black" #set the value globally
```

```python
fontsize_legend = 45
fontsize_title_legend = 45
```

## RF

```python
features = df[df.algorithm == "RF"].set_index('algorithm').T.sort_values(by=['RF'], ascending=False)
features = features[(features["RF"] > 0.01)]

names_features = []
for i in features.index:
    names_features.append(i[8:])
```

```python
fig, ax = plt.subplots(figsize=(40,20))
plot = sns.lineplot(data=features, linewidth=25, sort=False, ax=ax)
plot.set_xticklabels(names_features, rotation=45, horizontalalignment='right')
plot.set_xlabel('Caracter√≠stica', fontweight="bold")
plot.set_ylabel('Coeficiente de correlaci√≥n', fontweight="bold")
plot.get_legend().remove()

plt.savefig(plots_path + "features_rf.pdf")
```
